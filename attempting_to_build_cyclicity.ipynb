## Cyclic v periodic.
All periodic process are cyclic (go in cyclics) but not vice versa.
Non periodic cyclic processes have no time interval (P) such that a shift along the time axis by P leaves it invariant

## Trajectories and Invariants

How do we reparametrizate invariant functionals of trajectories.

Iterated integrals are explicitly reparamatrization invariant of a trajectory γ and defined in an inductive fashion. Iterated integration is the process of repeatedly integrating the results of previous integrations. Iterated intergrals of order zero are constants, associating to any trajectory γ a number <math xmlns="http://www.w3.org/1998/Math/MathML">

### Different orders

Orders decribe the order in which multiple integrals are evaluated. 

Order 1: the invariants of order 1 are exactly the increments of the trajectory over the interval I (basically just dxdy)
Order 2: renders more information than mere increments. Defined as: 
\begin{equation*}
\mathrm{I}_{\mathrm{k},\mathrm{l}}=\int_{\mathrm{t}_{\mathrm{s}}}^{\mathrm{t}_{\mathrm{f}}}\gamma_{\mathrm{k}}(\mathrm{t})\mathrm{d}\gamma_{\mathrm{l}}(\mathrm{t}).
\end{equation*}

and forms a vector space spanned by the alebraic areas encompassed by the projections of the trajectory γ onto various co-ordinate 2-planes (γ might need to be closed up by adding a striaght segment connecting the start and end point of the resultant line)

A key detail for this, is that cyclic phnomena is self-sustained cyclicing, where ane active process triggers another one etc (think one neuron firing to another one etc). So if one effect preceds another the oriented area on the parameteric joint plot of the corresponding functions will surround a positive algebraic area. This assumes that k = 1, if not then the lead-follow relationship flips.

- Oriented area is the surface area with a specified orientation (either positive to indicate going up, negative to indicate going down etc)
- A parametric plot is one in which a function or expression is plotted against another function or expression that uses the same independent variable

```python
def lead_lag_matrix_check(data):
    
    '''
    Function to calculate the lead lag
    matrix for cyclic analysis. 

    my function


    '''

    X = pd.DataFrame(data, dtype='float64')
    diff = X.diff()
    A = X[:-1].T @ diff.values[1:]
    return np.array(0.5 * (A - A.T))
```

```python
import numpy as np
import pandas as pd

def areaval(x: np.array, y: np.array) -> float:
    '''
    Function to Compute areavalue between two vectors.

    Parameters
    ----------
    x: np.array
        array of x values
    y: np.array 
        array of y values 

    Returns
    -------
    float: float number
        float value of area value
    '''
    return (np.dot(x, np.diff(np.concatenate((y[-1:], y)))) - np.dot(y, np.diff(np.concatenate((x[-1:], x))))) / 2

def lead_lag_matrix(arr):
    '''
    Function to make lead lag matrix

    Parameters
    ----------
    arr: np.array
        array of time series

    Returns
    -------
    lead_lag_matrix: np.array
        lead lag matrix
    '''
    N = arr.shape[1]
    lead_lag_matrix = np.zeros((N, N), dtype=arr.dtype)
    for index, index_col in enumerate(arr.T):
        for y_value in range(index+1, N):
            lead_lag_matrix[index, y_value] = areaval(index_col, arr[:, y_value])
            lead_lag_matrix[y_value, index] = -lead_lag_matrix[index, y_value]
    return lead_lag_matrix
```
