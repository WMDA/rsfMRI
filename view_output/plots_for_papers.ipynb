```python
import pandas as pd
from nilearn import plotting, datasets, image
```

```python
msdl = datasets.fetch_atlas_msdl()
msdl_overview_df = pd.DataFrame({'labels': msdl['labels'],
                                  'networks': msdl['networks'],
                                  'region_coords': msdl['region_coords']
                                  })
```

```python
img_index_autistic = [22, 23, 24, 37, 38]
img_index_class = [25, 26]
```

```python
classification_plotting= pd.DataFrame(
    data={
        'L STS':[0, 1, 0, 0],
        'R STS': [1, 0, 0, 0],
        'L IPS': [0, 0, 0, -0.5],
        'R Post Temp': [0, 0, -0.5, 0]
          },
    index=['L STS','R STS', 'L IPS','R Post Temp']
)
classification_regions = [[-52.12, -17.92, 13.28], [52.61, -13.65, 12.11], [-29.39, -59.43, 44.2], [62.53, -32.99, -9.14] ]
```

```python
plotting.plot_connectome(
    classification_plotting.values,
    classification_regions,
    title="A) Within (red) and Between (green) Networks Predictive Pairwise Realtionships",
    node_color='skyblue',
    node_size=50,
    edge_cmap='hsv',
    black_bg=True,
)
```

```python
plotting.plot_prob_atlas(
    image.index_img(
        msdl['maps'], 
        img_index_class,
        ),
    draw_cross=False,
    black_bg=True,
    cut_coords=[-52.12, -17.92, 13.28],
    cmap='gist_rainbow',
    title='B) Most Predictive Network'
)
```

```python
regression_dataframe = pd.DataFrame(
    data={
        'R Ant IPS':[0, 1, 0, 0, 0, 0, 0, 0],
        'L Ant IPS': [1, 0, 0, 0, 0, 0, 0, 0],
        'L IPS': [0, 0, 0, 1, 0, 0, 0, 0],
        'R Aud': [0, 0, 1, 0, 0, 0, 0, 0],
        # ADOS
        'V ACC':[0, 0, 0, 0, 0, -0.5, 0, 0],
        'D ACC': [0, 0, 0, 0, -0.5, 0, 0, 0],
        'Dors PCC': [0, 0, 0, 0, 0, 0, 0, -0.5],
        'L Front pol': [0, 0, 0, 0, 0, 0, -0.5, 0],
        
          },
    index=['R Ant IPS','L Ant IPS', 'L IPS','R Aud', 'V ACC', 'D ACC', 'Dors PCC', 'L Front pol' ]
)
regression_regions = [[48.36, -29.04, 43.13], [-47.85, -32.07, 41.9], [-29.39, -59.43, 44.2], [53.47, -6.49, 27.52], 
                      [-0.45, 34.06, 20.73], [-28.17, 46.32, 21.56], [-1.44, -59.12, 55.25], [-40.08, 50.65, 0.81]  ]
```

```python
plotting.plot_connectome(
    regression_dataframe.values,
    regression_regions,
    title="A) AQ10 (red) and ADOS-2 (green) Predictive Pairwise Realtionships",
    node_color='skyblue',
    node_size=50,
    edge_cmap='hsv',
    black_bg=True,
)
```
